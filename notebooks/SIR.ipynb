```python
import scipy.integrate
import numpy as np
import matplotlib.pyplot as plt
```

```python
ndays = 100
dt = 1 #time step in days
beta = (1.0/3.0) #infection rate
gamma = (1.0/14.0) #recovery rate

S = np.zeros(ndays) #susceptible
I = np.zeros(ndays) #infective
R = np.zeros(ndays) # removed
t = np.arange(ndays)*dt
```

```python
I[0] = 0.001 #initial infective proportion
S[0] = 1.0 - I[0] #initial susceptible
R[0] = 0.0
```

```python
for i in range(ndays-1):
  S[i+1] = S[i] - beta*(S[i]*I[i])*dt
  I[i+1] = I[i] + (beta*S[i]*I[i]-gamma*I[i])*dt
  R[i+1] = R[i] + (gamma*I[i])*dt
```

```python
fig = plt.figure(1); fig.clf()
plt.plot(t, S, 'r', lw=3, label="Susceptible")
plt.plot(t, I, 'g', lw=3, label="Infective")
plt.plot(t, R, 'b', lw=3, label="Removed")
fig.legend()
plt.xlabel('Days')
plt.ylabel('Fraction of Population')
```
