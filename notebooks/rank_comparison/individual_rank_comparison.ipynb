```python
import pandas as pd
import os
```

```python
input_path = "../rank_features/"
out_path = "./out/"
files = os.listdir(input_path)
step = 100

# ranks = ["bcr", "ccr", "page_rank", "user_rank", "sir", "ai", "aaj", "aaij"]
```

```python
def initialize_df():
    columns = ["Col"]
    [columns.append(f"range{i}") for i in range(1, 9)]
    df = pd.DataFrame(columns=columns)
    return df
```

```python
for r_name in ranks:
    file_group = [f for f in files if f.startswith(r_name)]
    out_df = initialize_df()

    for file_name in file_group:
    #     read df
        df = pd.read_csv(input_path+str(file_name))
        df.drop(df.columns[[0, 1]], axis = 1, inplace = True)

        for col in list(df.columns):
            col_data = df[col]
            n = len(col_data)
            print(col)
            c = 0

            data = [str(col)]

            for i in range(0, n-step, step):
                total = col_data[i:i+step].sum()
                avg = total/step
                c += 1
                print(f"range{c}: {avg}")
                data.append(avg)
            out_df.loc[len(out_df.index)] = data
            print("\n")
    out_df.to_csv(out_path+r_name+".csv", index=False)
```
