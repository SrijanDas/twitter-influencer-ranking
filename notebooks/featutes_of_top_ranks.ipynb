## Function to Fetch features

```python

def fetch_features(rank_df, rank_df_match_column_name, feat_df, feat_df_match_column_name, start_indx, end_indx, out_df_name="rank_features"):
    df = pd.DataFrame(columns=list(feat_df.columns.values))
    for idx, row in rank_df[start_indx:end_indx].iterrows():
        data = feat_df[feat_df[feat_df_match_column_name] == row[rank_df_match_column_name]]
        data = list(data.iloc[0])
        df.loc[len(df)] = data
    df.to_csv(f"./rank_features/{out_df_name}.csv", index=False)
    return df
```
