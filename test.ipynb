```python
import json
import networkx as nx
```

```python
network = {}
user_friends = {}
with open('.//output//Social Network_saved_data.json') as rf:
    network = json.load(rf)
    for i, j in network.items():
        user_friends[str(i)] = j['friends']
```

```python
user_friends
```

```python
g = nx.DiGraph()
g.add_nodes_from(user_friends.keys())
```

```python
g.order()
```

```python
for k, v in user_friends.items():
    g.add_edges_from(([(k, t) for t in v]))
```

```python
g.size()
```

```python
import matplotlib.pyplot as plt
```

```python
nx.draw(g)
```

```python
plt.show()
```

```python
 b=nx.betweenness_centrality(g,weight=None,normalized=True)
```

```python
c=nx.closeness_centrality(g)
```
